```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

## Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio

```python
# Crie um Barplot com o Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio
fig = plt.figure(figsize=(10,8))
print(df['fuelType'].count())

unique_gear = df['gearbox'].unique()
unique_gear_size = len(unique_gear)
palette = dict(zip(unique_gear, sns.color_palette(n_colors=unique_gear_size)))

sns.barplot(data=df, x=df['fuelType'], y=df['price'], hue=df['gearbox'], palette=palette)

plt.xlabel('Fuel Type', fontsize=18)
plt.ylabel('Average Price', fontsize=18)
plt.title('Average Price based on the Fuel Type and Transmission', fontsize=20)
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise3/fueltype-vehicleType-price.png", bbox_inches="tight")
```

## Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio

```python
# Crie um Barplot com a Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio
fig = plt.figure(figsize=(10,8))

sns.barplot(data=df, x=df['vehicleType'], y=df['powerPS'], hue=df['gearbox'], palette=["red","green","grey"])

plt.xlabel('Vehicle Type', fontsize=18)
plt.ylabel('Average Power', fontsize=18)
plt.title('Average Power based on the Vehicle Type and Transmission', fontsize=20)
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise3/vehicletype-fueltype-power.png")
```
