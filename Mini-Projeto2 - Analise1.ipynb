## Análise 1

```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

## Distribuição de Veículos com base no Ano de Registro

```python
# Crie um Plot com a Distribuição de Veículos com base no Ano de Registro
df['yearOfRegistration']
fig = plt.figure(figsize=(8,7))
ax = df['yearOfRegistration'].plot.hist(bins=50, color=('lightgreen'), alpha=0.8, density=True)

df['yearOfRegistration'].plot.kde(ax=ax, xlim=[1920,2020]) # O KDE pega cada registro e transforma em uma curva similar à Gaussiana (o pico da curva é o valor do registro. Ex.: ano do carro), somando todos entre si para obter uma única curva de KDE
#ax.set_xlim(xlim)
#plt.plot(df['yearOfRegistration'], range(len(df['yearOfRegistration'])))
plt.xlabel('Year of Registration', fontsize=18)
plt.ylabel('Density', fontsize=18)
plt.title('Vehicle Distribution based on Years of Registration', fontsize=20)
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise1/vehicle-distribution.png", bbox_inches='tight')
```

## Variação da faixa de preço pelo tipo de veículo

```python
# Crie um Boxplot para avaliar os outliers
print("Tipos =>", df['vehicleType'].values)
print("Tipos únicos =>", df['vehicleType'].unique())
print("Valores =>", df['price'].values)

fig = plt.figure(figsize=(8,7))

sns.set(style="whitegrid")
sns.boxplot(data=df, y=df['price'], x=df['vehicleType'], palette=["red"], width=0.7)

plt.xlabel('Vehicle Type', fontsize=18)
plt.ylabel('Price', fontsize=18)
plt.title('Outlier Analysis', fontsize=20)
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise1/price-vehicleType-boxplot.png", bbox_inches='tight')
```

## Contagem total de veículos à venda conforme o tipo de veículo

```python
# Crie um Count Plot que mostre o número de veículos pertencentes a cada categoria 
fig = plt.figure(figsize=(10,6))
print(df['vehicleType'].value_counts())
print(df['vehicleType'])
ax = sns.countplot(data=df, x=df['vehicleType'])
ax.bar_label(ax.containers[0])

plt.xlabel('Vehicle Type', fontsize=18)
plt.ylabel('Total Vehicles for Sale', fontsize=18)
plt.title('Total vehicle for sale counting based on each type', fontsize=20)
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise1/count-vehicleType.png", bbox_inches='tight')
```
