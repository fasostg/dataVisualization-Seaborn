```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
df.columns
```

```python
# Calcule a média de preço por marca e por veículo
df.groupby(['brand','vehicleType'])['price'].mean()
```

## Preço médio de um veículo por marca, bem como tipo de veículo

```python
# Crie um Heatmap com Preço médio de um veículo por marca, bem como tipo de veículo
fig = plt.figure(figsize=(14,12))

df_heat = pd.pivot_table(df, values='price', index=['brand'], columns=['vehicleType'], aggfunc=np.mean, fill_value=0) #pivot_table tem o objetivo de "remoldar" o Dataframe baseado nas colunas e índices que forem sugeridos
sns.heatmap(df_heat, cmap="Blues", linewidth=2, linecolor="lightgrey", annot=True, fmt='.0f')

plt.xlabel('Brand', fontsize=18)
plt.ylabel('Vehicle Type', fontsize=18)
plt.title('Heatmap - Average Price based on Vehicle Brand and Type', fontsize=20)
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise4/heatmap-price-brand-vehicleType.png")
```
