```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

## Número de veículos pertencentes a cada marca

```python
# Crie um Plot que mostre o número de veículos pertencentes a cada marca
fig = plt.figure(figsize=(11,8))

print(df['brand'])

sns.set_style("whitegrid", {'axes.axisbelow': True})
sns.countplot(data=df, y=df['brand'], palette='blend:darkblue,lightpink')


plt.xlabel('Vehicles Number', fontsize=18)
plt.ylabel('Brand', fontsize=18)
plt.title('Vehicles by Brand', fontsize=20)
plt.show()
```

```python
# Salvando o plot
fig.savefig(("plots/Analise2/brand-vehicleCount.png"), bbox_inches='tight')
```

## Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio

```python
# Crie um Plot com o Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio
fig = plt.figure(figsize=(10,8))
sns.barplot(data=df, x=df['vehicleType'], y=df['price'], hue=df['gearbox'])

plt.xlabel('Vehicle Type', fontsize=16)
plt.ylabel('Average Price', fontsize=16)
plt.title('Vehicles Average Price based on types and car transmission (manual, automatic and unknown)', fontsize=18)
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise2/vehicletype-gearbox-price.png", bbox_inches="tight")
```
